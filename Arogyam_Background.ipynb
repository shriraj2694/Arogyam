```python
import pandas as pd
import math

def calculate_combined_amount_of_carbon(row):
    thm_initial_concentration = row['Trihalomethanes']
    chloramine_initial_concentration = row['Chloramines']
    volume = 1 # litres
    thm_removal_efficiency = 0.90
    chloramine_removal_efficiency = 0.85

    thm_carbon = (thm_initial_concentration * volume) / (1 - thm_removal_efficiency) / 1000
    chloramine_carbon = (chloramine_initial_concentration * volume) / (1 - chloramine_removal_efficiency) / 1000
    combined_carbon = thm_carbon + chloramine_carbon

    return combined_carbon #kg

def calculate_filter_radius(combined_carbon, filter_height):
    # Assuming a cylindrical filter shape
    # Volume = pi * radius^2 * height
    # Solving for radius: radius = sqrt(Volume / (pi * height))
    radius = math.sqrt(combined_carbon / (math.pi * filter_height))
    return radius

def calculate_cost(combined_carbon, cost_per_kg):
    return combined_carbon * cost_per_kg

# Read the CSV file into a DataFrame
df = pd.read_csv('carbondata.csv')

# If 'Volume' column is available, replace 'Volume' with the actual column name for volume
# If 'Volume' column is not available, you may need to provide the volume information in another way

# Constants
filter_height = 2  # in feet
cost_per_kg = 600  # in Rs

# Calculate combined amount of carbon and store in a new column
df['Carbon_req'] = df.apply(calculate_combined_amount_of_carbon, axis=1)

# Calculate filter radius for each 'Carbon_req' value
df['Filter_radius'] = df.apply(lambda row: calculate_filter_radius(row['Carbon_req'], filter_height=filter_height), axis=1)

# Calculate cost for each 'Carbon_req' value
df['Cost'] = df.apply(lambda row: calculate_cost(row['Carbon_req'], cost_per_kg=cost_per_kg), axis=1)

# Display the DataFrame with the new 'Filter_radius' and 'Cost' columns
print("DataFrame with Filter_radius and Cost:")
print(df[['Carbon_req', 'Filter_radius', 'Cost']])


```
